* Remove Duplicates – Drop duplicate rows.
* Handle Missing Values – Remove or fill missing values.
* Lowercasing – Convert text to lowercase.
* Remove Special Characters and Punctuation – Remove unwanted symbols.
* Remove URLs – Eliminate links from tweets.
* Remove Mentions and Hashtags – Remove @username and #hashtags.

### Data Prepation

```python
import re
import string

# Manually defined stopwords list
english_stopwords = set([
    'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 
    'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', 'her', 
    'hers', 'herself', 'it', 'its', 'itself', 'they', 'them', 'their', 'theirs', 
    'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', 'these', 'those', 
    'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 
    'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 
    'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 
    'against', 'between', 'into', 'through', 'during', 'before', 'after', 'above', 
    'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 
    'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 
    'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 
    'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 
    's', 't', 'can', 'will', 'just', 'don', 'should', 'now'
])

def clean_tweet_text_with_numbers(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)  # Remove URLs
    text = re.sub(r'@\w+|\#', '', text)  # Remove mentions and hashtags
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    text = re.sub(r'[^\x00-\x7F]+', '', text)  # Remove non-ASCII characters like special symbols
    
    # Tokenize text and remove stopwords carefully
    words = text.split()
    words_cleaned = [word.strip() for word in words if word.strip() not in english_stopwords]
    text_cleaned = ' '.join(words_cleaned)

    return text_cleaned.strip()

# Apply the cleaning function to the 'tweet_text' column
df['cleaned_tweet'] = df['Text'].apply(clean_tweet_text_with_numbers)

# Display cleaned tweets
df[['Text', 'cleaned_tweet']].head()
df
```
